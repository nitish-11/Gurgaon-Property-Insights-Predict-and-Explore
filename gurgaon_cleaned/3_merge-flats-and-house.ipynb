```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
flats = pd.read_csv('../gurgaon_data/flats_cleaned.csv')
houses = pd.read_csv('../gurgaon_data/house_cleaned.csv')
```

```python
flats.sample(2)
```

```python
houses.sample(2)
```

```python
df = pd.concat([flats,houses], ignore_index=True)
```

```python
df.shape
```

- The code df.sample(df.shape[0], ignore_index=True) is used to randomly shuffle the rows of a DataFrame
- df.shape[0]: This gets the number of rows in the DataFrame df. For example, if df has 100 rows, df.shape[0] will return 100.

- df.sample(df.shape[0], ignore_index=True):

- df.sample(df.shape[0]): This randomly samples all rows of the DataFrame (i.e., the entire DataFrame) in a shuffled order.
- ignore_index=True: This resets the index of the shuffled DataFrame so that it starts from 0 without retaining the original index.
This results in a shuffled version of the DataFrame with the row indices reset.

```python
df = df.sample(df.shape[0],ignore_index=True)
```

```python
df.to_csv('../gurgaon_data/gurgaon_flats_houses.csv',index=False)
```
