df['society']:

- df refers to a Pandas DataFrame.
['society'] selects the column named 'society' from the DataFrame.
.apply(lambda name: ...):

- The .apply() function is a Pandas method that applies a function to each element of a series (in this case, the 'society' column).
lambda name: ... is an anonymous function that takes name as an argument. This name refers to each value in the 'society' column.
re.sub(r'\d+(\.\d+)?\s?★', '', ...):

- re.sub() is a function from the re (regular expression) module in Python. It replaces the part of the string that matches the pattern with the specified replacement ('' in this case, which removes the matched part).
- r'\d+(\.\d+)?\s?★' is the regular expression pattern:
- \d+ matches one or more digits (numbers).
- (\.\d+)? optionally matches a period (.) followed by one or more digits (to account for decimal numbers).
- \s? optionally matches any whitespace character.
- ★ matches the star symbol (★).

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['price'].dtype
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```
